##Renombre de columnas

Informacion de atributos:

Attribute Information:

This research employed a binary variable, default payment (Yes = 1, No = 0), as the response variable. This study reviewed the literature and used the following 23 variables as explanatory variables: 

* X1: Amount of the given credit (NT dollar): it includes both the 
individual consumer credit and his/her family (supplementary) credit. 

* X2: Gender (1 = male; 2 = female). 

* X3: Education (1 = graduate school; 2 = university; 3 = high school; 4 = others). 

* X4: Marital status (1 = married; 2 = single; 3 = others). 

* X5: Age (year). 

* X6 - X11: History of past payment. We tracked the past monthly payment records (from April to September, 2005) as follows: X6 = the repayment status in September, 2005; X7 = the repayment status in August, 2005; . . .;X11 = the repayment status in April, 2005. The measurement scale for the repayment status is: -1 = pay duly; 1 = payment delay for one month; 2 = payment delay for two months; . . .; 8 = payment delay for eight months; 9 = payment delay for nine months and above. 

* X12-X17: Amount of bill statement (NT dollar). X12 = amount of bill statement in September, 2005; X13 = amount of bill statement in August, 2005; . . .; X17 = amount of bill statement in April, 2005. 

* X18-X23: Amount of previous payment (NT dollar). X18 = amount paid in September, 2005; X19 = amount paid in August, 2005; . . .;X23 = amount paid in April, 2005. 

```python
#renombre de las columnas
df=df.rename(columns={'X1':'amount', 'X2':'gender', 'X3':'education', 'X4':'estado_civil', 'X5':'edad', 
                   'X6':'hist_sept','X7':'hist_ago', 'X8':'hist_jul', 'X9':'hist_jun', 'X10':'hist_may', 'X11':'hist_abr',
                   'X12':'bill_sept','X13':'bill_ago','X14':'bill_jul', 'X15':'bill_jun', 'X16':'bill_may', 'X17':'bill_abr',
                   'X18':'pago_sept', 'X19':'pago_ago', 'X20':'pago_jul', 'X21':'pago_jun','X22':'pago_may', 'X23':'pago_abr'})
df.head()
```

```python
#Convertir columnas a categoricas
categoricals = ['gender', 'education', 'estado_civil', 'hist_sept', 'hist_ago', 'hist_jul', 'hist_jun', 'hist_may', 'hist_abr']

df[categoricals] = df[categoricals].astype('object')
```

#3. Limpieza de datos 

Eliminando los registros nulos o rellena con la media de la columna

```python
#imputar categoricas con la moda

ndf = df.copy()

for cat in categoricals:
  mode = ndf[cat].mode()[0]
  ndf[cat].fillna(value = mode, inplace=True)
  print(cat,mode)

```

```python
#imputar continnuas con la media
continuas = list(e for e in ndf.columns if e not in categoricals)[1:-1]

for cont in continuas:
  mean = ndf[cont].mean()
  ndf[cont].fillna(value = mean, inplace=True)
  print(cont,mean)
```

```python
ndf.isna().sum()
```

Eliminamos los datos faltantes restantes en Y

```python
ndf.dropna(axis=0,how='any',inplace=True)
ndf.isnull().sum().sum() #comprobamos que ya no queden datos faltantes
```

# Reduccion de dimeciones con PCA

```python
from sklearn.decomposition import PCA
```

```python
#Seccionar componentes continuas para PCA
pca_ndf=ndf_sc[continuas]
```

```python
#PCA con variables escalas
pcs = PCA()
pcs.fit(pca_ndf)
pcsSummary_df = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                             'Proportion of variance': pcs.explained_variance_ratio_,
                             'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)
                             })

pcsSummary_df
```

```python
PC_components = np.arange(pcs.n_components_) + 1
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = pcs.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components-1, 
                 y = np.cumsum(pcs.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Acumulacion de varianza')
plt.xlabel('# Componente principal')
plt.ylabel('Varianza explicada')
plt.ylim(0, 1)
plt.show()
```

Observando la grafica, podemos determinar que con 5 componentes se logra explicar el 74% de la varianza de los datos continuos

```python
pcs = PCA(n_components=5)
pcs.fit(pca_ndf)
pcsSummary_df = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                             'Proportion of variance': pcs.explained_variance_ratio_,
                             'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)
                             })

pcsSummary_df
```

```python
PC_components = np.arange(pcs.n_components_) + 1
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = pcs.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components-1, 
                 y = np.cumsum(pcs.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Acumulacion de varianza')
plt.xlabel('# Componente principal')
plt.ylabel('Varianza explicada')
plt.ylim(0, 1)
plt.show()
```

## Importancia de las variables en cada componente

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(), 
                                columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5'],
                                index=[continuas])
pcsComponents_df
```

```python
#visualización Eigenvalores absolutos variables estandarizadas

plt.figure(figsize=(15,6))

variables = list(pcsComponents_df.index.get_level_values(0))

for c in pcsComponents_df.columns:
  plt.plot(variables, np.absolute(pcsComponents_df[c]),  marker='o', markersize=5, label=c)

plt.plot(variables, np.absolute(pcsComponents_df).sum(axis=1),  marker='o', markersize=5, label="Sumatoria")
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))

plt.title('Eigenvalores Absolutos escaladas')
plt.xlabel('Variable esclada')
plt.ylabel('Eigenvalor Absoluto')

plt.xticks(rotation=90)
plt.show()
```
