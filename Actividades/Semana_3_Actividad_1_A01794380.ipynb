#Parte 1: Fundamentos de bases de datos
___
1. **Fundamentos de bases de datos y para ciencia de datos.**

Una base de datos es un set estructurado de información que se encuentra almacenado en una computadora o en la nube. 
Los datos dentro de la base de datos están organizados en forma de tablas. Estas tablas están constituidas por filas (renglones) que representan cada entrada o registro y por columnas, donde las columnas son las características o campos de cada registro.

En general, las bases de datos se clasifican en dos tipos: Relacionadas y No relacionadas.
Las bases relacionadas consisten en tablas que están conectadas entre ellas por algún tipo de relación o característica común entre las tablas.

Las bases no relacionadas consisten datos almacenados en forma con un estilo de documentos, no son tabulares o tablas como las previas mencionadas. La ventaja de este tipo de base de datos es que son óptimas para datos dinámicos que cambian o se actualizan frecuentemente, por lo que son más eficientes, como por ejemplo datos de seguro médico de pacientes.

El objetivo principal de una base de datos, es el de modelar algún sistema particular mediante el registro de las características que permitan 
representarlo de forma precisa con datos en la computadora.

___

2. **Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.**

Un almacén de datos (data warehouse) es un sistema de administración de información integrado por varios elementos que ayudan a administrar grandes cantidades de información de forma eficiente y consolidándola en un solo sistema.

Alguno de los componentes incluye distintos tipos de bases de datos relacionadas, sistemas de extracción, carga y transformación (ELT por sus siglas en inglés) para preparar los datos, sistemas de análisis estadístico y minería, entre otros.

## Solución 1: Descartar las observaciones con valores faltantes - Eliminación por lista

### Por renglón
Eliminamos todos las observaciones que tienen por lo menos algún dato perdido y lo guardamos en un nuevo DF

```python
df1 = df.dropna()

print ('DF sin datos perdidos:')
df1.isnull().sum()
```

Comparamos el tamaño del DF original vs el nuevo

```python
print('# datos en DF original: ', df.shape[0])
print('# datos en DF nuevo: ', df1.shape[0])
print('Total de datos eliminados: ', df.shape[0]-df1.shape[0])
```

### Por columna
Se elimina la columa completa si tiene por lo menos algún dato perdido y lo guardamos en un nuevo DF

```python
df2 = df.dropna(axis = 1) # axis 1 is columns / axis 0 is rows. 
print ('DF sin datos perdidos:')
df2.isnull().sum()
```

Con el método de eliminación por columna, perdemos 22 columnas (de 23) lo cual no es un método valido para este set de datos 

###Eliminación por parejas



Alternativamente: usamos Threshold.

Mentenemos solo las filas con al menos 2 valores que NO SEAN `nan`

```python
df3 = df.dropna(thresh=22) #Son 25 columnas, el corte requiere que existan 22 columnas sin NAN en cada linea
print ('DF con datos perdidos filtrados:')
df3.isnull().sum()
```

```python
print('# datos en DF original: ', df.shape[0])
print('# datos en DF nuevo: ', df3.shape[0])
print('Total de datos eliminados: ', df.shape[0]-df3.shape[0])
```

```python
df3[df3.isnull().any(axis=1)].sample(3)
```

```python
df4 = df.dropna(thresh=29990, axis=1) #Son 30000 lineas, el corte requiere que NO existan mas de 10 lineas NAN en cada columna, si no, se elimina la columna
print ('DF con datos perdidos filtrados:')
df4.isnull().sum()
```

## Solución 2: Sustitución - Media

Calculamos la media de la columna X2

```python
x16_m = df.X16.mean()
x16_m
```

```python
df5 = df.copy()

df5['X16'].fillna(value = x16_m, inplace=True)

print ('DF con datos perdidos filtrados:')
df5.isnull().sum()
```

```python
df.describe().T
```

### Moda
Imputar datos perdidos con la moda

```python
df6 = df.copy()

for columna in df6.columns[2:]:
  df6[columna].fillna(df6[columna].mode()[0], inplace=True)
  print(columna, df6[columna].mode()[0])

print ('DF con datos perdidos filtrados:')
df6.isnull().sum()
```

```python
df6.describe().T
```

### Mediana
Imputar datos perdidos con la moda en Variables X

Eliminar datos faltantes de Y

```python
df7 = df.copy()

for columna in df7.columns[2:23]:
  df7[columna].fillna(df7[columna].median(), inplace=True)
  print(columna, df7[columna].median())

df7.dropna(inplace=True)

print ('DF con datos perdidos filtrados:')
df7.isnull().sum()
```

```python
df7.describe().T
```
